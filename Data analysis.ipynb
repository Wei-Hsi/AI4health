```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/Hsuweic/Desktop/AI4healthcare/dataset/dataset_categories.csv')

# Set a darker shade of blue
dark_blue = sns.color_palette('deep')[0]

# Plotting the data distribution for 'label'
plt.figure(figsize=(8, 8))
plt.title('Data Distribution', fontweight='bold', fontsize=14)
plt.xlabel(' ')  # Set x-axis label to an empty string
plt.ylabel('tweets')
plt.xticks(ticks=[0, 1], labels=['Non Hate Speech', 'Hate Speech'])  # Customize x-axis labels

# Annotate count values on top of each bar
ax = sns.countplot(x='label', data=df, palette=[dark_blue, 'lightblue'], width=0.4)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.legend(loc='upper right', labels=['Non Hate Speech', 'Hate Speech'])
plt.show()

# Data distribution for 'categories' column for hate speech (label=1)
hate_speech_df = df[df['label'] == 1]
categories_counts = hate_speech_df['categories'].value_counts()

# Plotting the data distribution for 'categories' for hate speech
plt.figure(figsize=(10, 10))
plt.title('Categories for Hate Speech')
plt.xlabel(' ')
plt.ylabel('tweets')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Annotate count values on top of each bar
ax = sns.countplot(x='categories', data=hate_speech_df, order=categories_counts.index, color=dark_blue, width=0.8)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Set a darker shade of blue
dark_blue = sns.color_palette('deep')[0]

# Data distribution for 'categories' column for hate speech (label=1)
hate_speech_df = df[df['label'] == 1]
categories_counts = hate_speech_df['categories'].value_counts()

# Plotting a pie chart for all categories with actual count and percentage
plt.figure(figsize=(12, 12))
plt.pie(categories_counts, labels=categories_counts.index,
        autopct=lambda p: '{:.0f} ({:.1f}%)'.format(p * sum(categories_counts) / 100, p),
        textprops=dict(fontsize=16), 
        colors=sns.color_palette('pastel'))

plt.title('Categories of Hate Speech', fontweight='bold', fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a darker shade of blue
dark_blue = sns.color_palette('deep')[0]

# Data distribution for 'categories' column for hate speech (label=1)
hate_speech_df = df[df['label'] == 1]
categories_counts = hate_speech_df['categories'].value_counts()

# Plotting a pie chart for all categories with actual count and percentage
plt.figure(figsize=(12, 12))
wedges, texts, autotexts = plt.pie(categories_counts, labels=categories_counts.index,
                                   autopct='',
                                #    autopct=lambda p: '{:.0f} ({:.1f}%)'.format(p * sum(categories_counts) / 100, p),
                                   textprops=dict(fontsize=16),  # Set text color and size
                                   colors=sns.color_palette('pastel'))

# Add legend
legend_labels = [f'{label} ({count})' for label, count in zip(categories_counts.index, categories_counts)]
plt.legend(wedges, legend_labels, title='Categories', loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Categories of Hate Speech', fontweight='bold', fontsize=20)  # Increase the font size of the title
plt.show()
```
